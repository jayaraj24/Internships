# Importing the libraries and reading the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', None)
%matplotlib inline
```

```python
df = pd.read_csv("Financial analytics data.csv",index_col=0)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.rename(columns={"Mar Cap - Crore":"Mar cap","Sales Qtr - Crore":"Q sales","Unnamed: 4":"New"},inplace=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

# Cleaning the data and Exploratory data Analysis

```python
df[df['Mar cap'].isnull()]
```

```python
df.drop(index=df[df['Mar cap'].isnull()].index,inplace = True)
df.isnull().sum()
```

```python
df.shape
```

```python
df[df['Q sales'].isnull()]
```

```python
nan_index = df[df['Q sales'].isnull()].index
nan_index
```

```python
for i in nan_index:
    if pd.isna(df.loc[i,"Q sales"]):
        df.loc[i,"Q sales"] = df.loc[i,"New"]
 
df.loc[nan_index]
```

```python
df.drop("New",axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df[df['Q sales'].isnull()]
```

```python
df.drop(index=df[df['Q sales'].isnull()].index,inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
df.head(10)
```

# Relationship between Market cap and Quarterly sales 

```python
plt.figure(figsize=(7,5))
plt.plot(df['Mar cap'],df['Q sales'],"og")

plt.xlabel("Market cap")
plt.ylabel("Quarterly sales")
plt.title("Relationship between Market cap and Quarterly sales")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.jointplot(data=df,x='Mar cap',y='Q sales',kind='reg') # linear relation 
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df,color="orange")
```

# Central tendancy and dispersion of the data

```python
CD_data = {"Market cap":[df["Mar cap"].mean(),df["Mar cap"].median(),df["Mar cap"].std(),df["Mar cap"].var()],"Quarterly sales":[df["Q sales"].mean(),df["Q sales"].median(),df["Q sales"].std(),df["Q sales"].var()]}

CD = pd.DataFrame(CD_data,index=['Mean',"Median",'Std','Variance'])
CD
```

```python
plt.figure(figsize=(5,5))
sns.barplot(data=CD)
```

# The Distribution of the data 

```python
fig,ax = plt.subplots(1,2,figsize=(10,4),width_ratios=[2,1.5])
plt.subplots_adjust(wspace=0.5)

sns.histplot(data=df['Mar cap'],bins=5,ax=ax[0],color='g')
sns.histplot(data=df['Q sales'],bins=5,ax=ax[1],color='r')
plt.show()
```
