```python
attrition_count = df['Attrition'].value_counts()
attrition_rate = attrition_count['Yes'] / len(df) * 100  # Convert to percentage

# Visualize attrition rate
plt.figure(figsize=(8, 6))
sns.barplot(x=attrition_count.index, y=attrition_count.values)
plt.title('Attrition Rate')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.text(0, attrition_count['Yes'], f'Attrition Rate: {attrition_rate:.2f}%', ha='center')
plt.show()
```

```python
# Visualize job satisfaction for employees who left vs. stayed
plt.figure(figsize=(10, 6))
sns.histplot(data=yes, x='JobSatisfaction', color='red',label='Attrition: Yes',discrete=True,stat='probability',shrink=0.8)
sns.histplot(data=no, x='JobSatisfaction', color='blue',label='Attrition: No',discrete=True,stat='probability',shrink=0.8)
plt.title('Job Satisfaction Distribution by Attrition Status')
plt.xlabel('Job Satisfaction')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Salary Distribution by Attrition Status')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Years at the company
plt.subplot(2, 2, 1)
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Years at Company by Attrition Status')

# Years since last promotion
plt.subplot(2, 2, 2)
sns.boxplot(x='Attrition', y='YearsSinceLastPromotion', data=df)
plt.title('Years Since Last Promotion by Attrition Status')

# Years with current manager
plt.subplot(2, 2, 3)
sns.boxplot(x='Attrition', y='YearsWithCurrManager', data=df)
plt.title('Years With Current Manager by Attrition Status')

plt.tight_layout()
plt.show()
```

```python
factors = ['WorkLifeBalance', 'EnvironmentSatisfaction', 'JobInvolvement']

# Create subplots to visualize each factor
fig, axes = plt.subplots(nrows=1, ncols=len(factors), figsize=(15, 5))

# Loop through each factor and create box plots
for i, factor in enumerate(factors):
    sns.boxplot(x='Attrition', y=factor, data=df, ax=axes[i])
    axes[i].set_title(f'{factor} by Attrition Status')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Box plot for TrainingTimesLastYear
sns.boxplot(x='Attrition', y='TrainingTimesLastYear', data=df, ax=axes[0])
axes[0].set_title('Training Times Last Year by Attrition Status')

# Box plot for PerformanceRating
sns.boxplot(x='Attrition', y='PerformanceRating', data=df, ax=axes[1])
axes[1].set_title('Performance Rating by Attrition Status')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
demographic_variables = ['Age', 'Gender', 'MaritalStatus', 'Education', 'JobRole']

fig, axes = plt.subplots(nrows=len(demographic_variables), ncols=1, figsize=(10, 8), sharex=True)

for i, variable in enumerate(demographic_variables):
    sns.countplot(x=variable, hue='Attrition', data=df, ax=axes[i])
    axes[i].set_title(f'Attrition by {variable}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
